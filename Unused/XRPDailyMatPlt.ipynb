```python
import pandas as pd
import os
```

```python
XRPDaily = os.path.join('Resources/XRPd.csv')
XRPd_df = pd.read_csv(XRPDaily)
XRPd_df
```

```python
XRPd_df['Date'] = pd.to_datetime(XRPd_df['Date'])
XRPd_df['Day'] = XRPd_df['Date'].dt.day_name()
XRPd_df
```
