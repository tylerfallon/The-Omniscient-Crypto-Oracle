```python
import pandas as pd
import os
```

```python
XRPDaily = os.path.join('Resources/XRPd.csv')
XRPd_df = pd.read_csv(XRPDaily)
XRPd_df
```

```python
XRPd_df['Date'] = pd.to_datetime(XRPd_df['Date'])
XRPd_df['Day'] = XRPd_df['Date'].dt.day_name()
XRPd_df
```

```python
XRPd_df.dtypes
```

```python
for column in XRPd_df.columns:
    print(f"Column {column} has {XRPd_df[column].isnull().sum()}null values")
```

```python
# Remove customer ID column
XRPd_df.drop(columns=["Open", "High", "Low", "Vol.", "Change %", "Date"], inplace=True)
XRPd_df.head()
```

```python
XRPd_df['Day1'] = XRPd_df.loc[:, 'Day']
XRPd_df
```

```python
# Transform string column
def change_string(Day1):
    if Day1 == "Monday":
        return 1
    if Day1 == "Tuesday":
        return 2
    if Day1 == "Wednesday":
        return 3
    if Day1 == "Thursday":
        return 4
    if Day1 == "Friday":
        return 5
    if Day1 == "Saturday":
        return 6
    else:
        return 0
    
XRPd_df["Day1"] = XRPd_df["Day1"].apply(change_string)
XRPd_df.head()
```

```python
XRPDay_df = XRPd_df.groupby(by='Day').mean().sort_values(by= "Day1")
XRPDay_df
```

```python
# Saving cleaned data
file_path = "Resources/XRPDay.csv"
XRPd_df.to_csv(file_path, index=False)
```
