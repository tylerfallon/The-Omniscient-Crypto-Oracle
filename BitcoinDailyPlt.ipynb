```python
import pandas as pd
import os
```

```python
BTCDaily = os.path.join('Resources/BitcoinDataDaily.csv')
BTCd_df = pd.read_csv(BTCDaily)
BTCd_df
```

```python
BTCd_df['Date'] = pd.to_datetime(BTCd_df['Date'])
BTCd_df['Day'] = BTCd_df['Date'].dt.day_name()
BTCd_df
```

```python
BTCd_df.dtypes
```

```python
#convert revenue column to float
BTCd_df['Price'] = BTCd_df['Price'].apply(lambda x: float(x.split()[0].replace(',', '')))

#view updated DataFrame
print(BTCd_df)
```

```python
BTCd_df['Price'] = BTCd_df['Price'].astype(float)
```

```python
BTCd_df.dtypes
```

```python
for column in BTCd_df.columns:
    print(f"Column {column} has {BTCd_df[column].isnull().sum()}null values")
```

```python
# Remove customer ID column
BTCd_df.drop(columns=["Open", "High", "Low", "Vol.", "Change %", "Date"], inplace=True)
BTCd_df.head(10)
```

```python
BTCd_df['Day1'] = BTCd_df.loc[:, 'Day']
BTCd_df
```

```python
# Transform string column
def change_string(Day1):
    if Day1 == "Monday":
        return 1
    if Day1 == "Tuesday":
        return 2
    if Day1 == "Wednesday":
        return 3
    if Day1 == "Thursday":
        return 4
    if Day1 == "Friday":
        return 5
    if Day1 == "Saturday":
        return 6
    else:
        return 0
    
BTCd_df["Day1"] = BTCd_df["Day1"].apply(change_string)
BTCd_df.head()
```

```python
BTCDay_df = BTCd_df.groupby(by='Day').mean().sort_values(by= "Day1")
BTCDay_df
```

```python
# Saving cleaned data
file_path = "Resources/BTCDay.csv"
BTCd_df.to_csv(file_path, index=False)
```
